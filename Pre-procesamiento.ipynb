```python
cur=con.cursor()
query = """
SELECT EmployeeID, strftime('%Y', retirementDate) AS retirementYear
FROM retirement
WHERE retirementDate IS NOT NULL;
"""

# Ejecutar la consulta
cur.execute(query)

# Recuperar los resultados
results = cur.fetchall()

# Mostrar los resultados
print("EmployeeID | Retirement Year")
print("-----------------------------")
for row in results:
    print(f"{row[0]:<10} | {row[1]}")

# Cerrar la conexión
con.close()
```

```python
df_consul = pd.read_sql("""
SELECT 
    a.*,  
    b.EnvironmentSatisfaction, 
    b.JobSatisfaction, 
    b.WorkLifeBalance,
    c.JobInvolvement, 
    c.PerformanceRating, 
    c.SurveyDate,
    d.EmployeeID,
    d.Attrition,
    d.retirementDate,
    d.retirementType,
    d.resignationReason
    
FROM 
    generaldata a 
LEFT JOIN 
    employeesurvey b 
ON 
    a.EmployeeID = b.EmployeeID AND a.InfoDate = b.DateSurvey 
LEFT JOIN 
    managersurvey c 
ON 
    a.EmployeeID = c.EmployeeID AND a.InfoDate = c.SurveyDate
LEFT JOIN 
    retirement d 
ON 
    a.EmployeeID = d.EmployeeID
""", con)
```
